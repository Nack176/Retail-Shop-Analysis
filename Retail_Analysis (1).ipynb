## Analysising Revenue in different form.

```python
# Add a Revenue column
df['Revenue'] = df['Price'] * df['Quantity']
# I add this Revenue Column to analysis on Revenue.
# Convert date columns to datetime
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df['Customer_Since'] = pd.to_datetime(df['Customer_Since'])

# Summary statistics
df.describe()
```

```python
# Monthly revenue trend
monthly_revenue = df.groupby(df['Purchase_Date'].dt.to_period('M'))['Revenue'].sum()
monthly_revenue.index = monthly_revenue.index.to_timestamp()

# Plot
plt.figure(figsize=(12, 5))
monthly_revenue.plot(marker='o', color='blue')
plt.title("Monthly Revenue Trend - Seasonality")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### The Above Figure showing the Monthly Revenue Trend.

```python
# Group by day
daily_orders = df.groupby(df['Purchase_Date'].dt.date).size()
daily_revenue = df.groupby(df['Purchase_Date'].dt.date)['Revenue'].sum()

# Combine into a DataFrame
daily_data = pd.DataFrame({
    'Order Volume': daily_orders,
    'Revenue': daily_revenue
})

# Plot
plt.figure(figsize=(14, 6))
plt.plot(daily_data.index, daily_data['Order Volume'], label='Order Volume', alpha=0.7)
plt.plot(daily_data.index, daily_data['Revenue'], label='Revenue', alpha=0.7)
plt.title("Daily Order Volume vs Revenue")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### The Above Figure showing the Daily Ordered volume VS Revenue Trend.
